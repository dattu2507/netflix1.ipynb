#Deleting unneeded columns.


*   Renaming the columns
*   Dropping the duplicates
*   Removing the NaN values from the data set







```python
#reading and checking the unneeded columns
data.columns
```

```python
#reading unneeded column
data[['rating']]
```

```python
#dropping unneeded column
data.drop(columns='rating',inplace=True)
```

```python
data.head()
```

```python
#reading columns after deletng unneeded
data.columns
```

```python
#changing column names
new_column_names=[]
for i in data.columns:
  print(i.capitalize())
  new_column_names.append(i.capitalize())
```

```python
#assigning new column names
data.columns=new_column_names
```

```python
data.head()
```

```python
data.columns
```

```python
#checking for duplicated values
data.duplicated().sum()
```

```python
#if there are any duplicate values
data.drop_duplicates(inplace=True)
```

```python
#checking for NaN values
data.isna().sum()
```

```python
#if there any NaN values
data.dropna(inplace=True)
```

```python
data['Show_id'].unique()
```

```python
#removing 's' for all elements
data['Show_id']=data['Show_id'].apply(lambda x:x.split('s')[1])
```

```python
data['Show_id']
```

```python
data
```

```python
#converting type of Show_id
data['Show_id']=data['Show_id'].astype(int)
```

```python
type(data['Show_id'][0])
```

```python
#replcing "/" with "-"
data['Date_added']=data['Date_added'].apply(lambda x:x.replace('/','-'))
```

```python
data
```

```python
data.to_csv('netflix_data.csv',index=False)
```
